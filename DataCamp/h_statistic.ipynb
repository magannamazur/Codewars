## Metoda gradientu w regresji

```python
import pandas as pd
punkty = list(pd.read_csv('https://bit.ly/2KF29Bd').itertuples())
```

```python
punkty
```

```python
m = 0.0
b = 0.0
L = 0.001
iteracje = 100000
n = float(len(punkty))
n
```

```python
# y = mx + b
# metoda sumy kwadratw

for i in range(iteracje):
    d_m = sum(2*p.x *((m*p.x+b)-p.y) for p in punkty)
    d_b = sum(2*((m*p.x+b)-p.y) for p in punkty)
    m -= L*d_m
    b -= L*d_b
```

# Stochastic gradient descent
- uses batches for learning in each iteration (diving the data into baches)

```python
import pandas as pd
import numpy as np

dane = pd.read_csv('https://bit.ly/2KF29Bd', header = 0)
X = dane.iloc[:,0].values
Y = dane.iloc[:,1].values
print(X)
print(Y)
```

```python
n = dane.shape[0]
m = 0.0
b = 0.0
rozmiar_proby = 1 # if more than 1, it iwl use nregular gradient method
L = 0.0001
epoki = 1_000_000
```

```python
for i in range(epoki):
    idx = np.random.choice(n, rozmiar_proby, replace = False)
    x_proba = X[idx]
    y_proba = Y[idx]
    Y_pred = m*x_proba + b
    d_m = (-2/rozmiar_proby)*sum(x_proba*(y_proba-Y_pred))
    d_b = (-2/rozmiar_proby)*sum(y_proba-Y_pred)
    m = m-L*d_m
    b = b-L*d_b
    if i % 10000 == 0:
        print(i,m,b)
```

```python
print('y = {0}x + {1}'.format(m,b))
```

## Przedzia ufnoci w regresji liniowej
Musimy znale margines bdu i doda/odj od przewidywanej wartoci y

```python
from PIL import Image 
from IPython.display import display
img = Image. open('przedzial.jpg')
display(img)
```

```python
punkty = list(pd.read_csv('https://bit.ly/2KF29Bd').itertuples())

n = len(punkty)

m = 1.939
b = 4.733

x_0 = 8.5
x_srednia = sum(p.x for p in punkty) / len(punkty)

wartosc_t = t(n - 2).ppf(.975)

blad_standardowy = sqrt(sum((p.y - (m * p.x + b)) ** 2 for p in punkty) / (n - 2))

margines_bledu = wartosc_t * blad_standardowy * sqrt(1+(1/n)+(n*(x_0-x_srednia)**2)/(n*sum(p.x**2 for p in punkty)\
                                                                                  -sum(p.x for p in punkty)**2))
przewidywane_y = m*x_0 + b
print(przewidywane_y-margines_bledu,przewidywane_y+margines_bledu,przewidywane_y)
```

by u weterynarza 21,2145 razy), ale moemy rwnie powiedzie co znacznie mniej absolutnego:

```python
#. Scale-location plot
# this plot shows whether the size of the residuals gets bigger or smaller.
# In a good model, the size of the residuals shouldn't change much as the fitted values change.
ms_vs_leng_residuals = ms_vs_leng.get_influence().resid_studentized_internal
ms_vs_leng_residuals_abs = np.sqrt(np.abs(ms_vs_leng_residuals))
sns.regplot(x =ms_vs_leng.fittedvalues, y = ms_vs_leng_residuals_abs, ci = None, lowess = True )
plt.xlabel('Fitted values')
plt.ylabel('Sqrt of abs val of stdized residuals')
```
