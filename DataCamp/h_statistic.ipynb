### The binomial distribution 
- it can tell us the probability of getting some number of heads in a sequence of coin flips

```python
from scipy.stats import binom
b1 =binom.rvs(1, 0.5, size = 8)
b1
# binom.rvs(how may coin, p = probability, n = how many time)
```

```python
binom.rvs(8, 0.5, size = 1)
# total numer of success of 8 coins
```

```python
binom.rvs(3, 0.5, size = 10)
```

```python
# To get the probability of getting 7 heads out of 10 
#binom.pmf(num of heads,num trial, probability of getting head)
binom.pmf(7,10,0.5)
```

```python
# the probability of getting a number of successes less than or equal to the first argument
binom.cdf(7,10,0.5)
```

## Mnozenie macierzy matmu(), @

```python
i3 = np.array([0,1])
j3 = np.array([-1,0])
baza3 = np.array([i3,j3])
baza3_t = baza3.transpose()
baza3_t
```

```python
i4 = np.array([1,0])
j4 = np.array([1,1])
baza4 = np.array([i4,j4])
baza4_t = baza4.transpose()
baza4_t
```

```python
po = baza4_t@baza3_t
po
```

```python
v4 = np.array([1,2])
po.dot(v4)
```

```python
po2 = baza3_t@baza4_t
po2
```

```python
v5 = np.array([1,2])
po2.dot(v5)
```

## Ukady rwna

```python
# 4x + 2y + 4z = 44
# 5x + 3y + 7z = 56
# 9x + 3y + 6z = 72
```

```python
from sympy import *
A = Matrix([[4,2,4],
          [5,3,7],
          [9,3,6]])
A
```

```python
B = Matrix([45,56,72])
B
```

```python
odwrotna = A.inv()
odwrotna
```

```python
jednoskowa = odwrotna*A
jednoskowa
```

```python
X = A.inv()*B
X
```

```python
from numpy.linalg import inv
A = np.array([[4,2,4],
          [5,3,7],
          [9,3,6]])
B = np.array([45,56,72])
X = inv(A).dot(B)
X
```
