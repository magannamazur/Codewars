# Bar
- show the mean of a quantitative variable among observations in each category
- automatically shows 95% confidence intervals for these means

```python
sns.catplot(x ='day', y = 'total_bill', data = tips, kind = 'bar')
plt.show()
```

```python
# turning off the confidence intervals
sns.catplot(x ='day', y = 'total_bill', data = tips, kind = 'bar', ci = None)
plt.show()
```

```python
sns.catplot(x ='total_bill', y = 'day', data = tips, kind = 'bar', ci = None)
plt.show()
```

```python
sns.catplot(x ='time', y = 'total_bill', data = tips, kind = 'box')
plt.show()
```

```python
sns.catplot(x ='time', y = 'total_bill', data = tips, kind = 'box', order = ['Dinner', 'Lunch'])
plt.show()
```

```python
# to omit the outliers from your box plot
sns.catplot(x ='time', y = 'total_bill', data = tips, kind = 'box', sym = '')
plt.show()
```

```python
# Changing the whiskers using `whis`
sns.catplot(x ='time', y = 'total_bill', data = tips, kind = 'box', whis = 2)
plt.show()
```

```python
# Changing the whiskers using `whis`
sns.catplot(x ='time', y = 'total_bill', data = tips, kind = 'box', whis = [5,95])
plt.show()
```

# Point plots
- show the mean of a quantitative variable for the observations in each category
- The vertical bars extending above and below the mean represent the 95% confidence intervals for that mean

```python
from PIL import Image 
from IPython.display import display
img = Image. open('point.jpg')
display(img)
```

```python
# Changing the whiskers using `whis`
sns.catplot(x ='time', y = 'total_bill', data = tips, kind = 'point', hue ='sex' )
plt.show()
```

```python
sns.catplot(x ='time', y = 'total_bill', data = tips, kind = 'point', hue ='sex', join = False )
plt.show()
```

```python
# Displaying the median
from numpy import median
sns.catplot(x ='smoker', y = 'total_bill', data = tips, kind = 'point', estimator = median)
plt.show()
```

```python
# Customizing the confidence intervals
sns.catplot(x ='smoker', y = 'total_bill', data = tips, kind = 'point', capsize = 0.05)
plt.show()
```

```python
sns.catplot(x ='smoker', y = 'total_bill', data = tips, kind = 'point', ci = None)
plt.show()
```

```python
ff = sns.catplot(x ='day', y = 'total_bill', data = tips, kind = 'bar', col = 'smoker')
plt.xticks(rotation = 90)
ff.fig.suptitle('totall bill', y = 1.03)
ff.set_titles('Subplot Smoker: {col_name}')
plt.xticks(rotation = 90)
plt.show()
```
