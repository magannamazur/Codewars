```python
def cv2_imshow(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
```

# Zadanie 1. Modyfikacja rozdzielczości

1. Wczytać obraz.
2. Zmodyfikować jego rozdzielczość do 256x256.
3. Zmodyfikować rozdzielczość oryginalnego obrazu do 40% (można na dwa sposoby - patrz dokumentacja).
4. Zmodyfikować rozdzielczość oryginalnego obrazu do szerokości 256 bez zniekształceń.

```python
img = cv2.imread('ML-datasets/images/gory_lc.png')
```

```python
img.shape
```

```python
cv2_imshow(img)
```

```python
# modyfikacja do wskazanej wielkości
resized = cv2.resize(img, dsize=(256, 256))
```

```python
cv2_imshow(resized)
```

```python
# modyfikacja do konkretnego % (bez zniekształceń)
```

```python
scale_percent = 40 # percent of original size

width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)

resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
```

```python
cv2_imshow(resized)
```

```python
# korzystając z funkcji wbudowanej - należy podać tylko parametr
```

```python
resized = cv2.resize(img, fx=0.4, fy=0.4, dsize=None)
```

```python
cv2_imshow(resized)
```

```python
# matematyczny sposób na znalezienie wysokości dla zadanej szerokości
```

```python
new_width = 30 # nasza docelowa szerokość
pct = new_width / img.shape[1]  # liczymy skalę dla szerokości

new_height = int(img.shape[0] * pct)  # na podstawie skali - nowa wysokość
```

```python
new_height
```

```python
dim = (new_width, new_height)

resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
```

```python
cv2_imshow(resized)
```
