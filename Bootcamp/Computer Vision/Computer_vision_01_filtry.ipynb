```python
def cv2_imshow(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
```

# Filtry

```python
import numpy as np

def apply_filter(source, kernel):
    new_image = cv2.filter2D(source, -1, kernel)
    cv2_imshow(new_image)
    return new_image

cv2_imshow(img)
```

```python
identity = np.array([[0,0,0],
                     [0,1,0],
                     [0,0,0]])
result = apply_filter(img, identity)
```

```python
blur = np.ones((3,3), np.float32) / 9
result = apply_filter(img, blur)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
identity_2 = np.array([[-1,-1,-1],
                     [-1,8,-1],
                     [-1,-1,-1]])
result = apply_filter(gray, identity_2)
```

```python
bottom_sobel = np.array([[-1,-2,-1],
                     [0,0,0],
                     [1,2,1]])
left_sobel = np.array([[1,0,-1],
                     [2,0,-2],
                     [1,0,-1]])
result_1 = apply_filter(gray, bottom_sobel)
result_2 = apply_filter(gray, left_sobel)

sobel = result_1 + result_2
cv2_imshow(sobel)
```
