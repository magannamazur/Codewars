```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import cv2
```

```python
img = cv2.imread('ML-datasets/images/lena.png')
```

```python
def threshold(image):
  @interact(x=widgets.IntSlider(min=0, max=255, step=1, value=100))
  def trackbar(x):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    ret, thresh_1 = cv2.threshold(gray, x, 255, cv2.THRESH_BINARY)
    cv2_imshow(thresh_1)

threshold(img)
```

```python
book = cv2.imread('ML-datasets/images/book_threshold.png')
```

```python
cv2_imshow(book)
```

```python
threshold(book)
```

```python
def adaptive_threshold(image):
  @interact(x=widgets.IntSlider(min=3, max=15, step=2, value=9), y=True)
  def trackbar(x, y):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    if y:
      method = cv2.ADAPTIVE_THRESH_MEAN_C
    else:
      method = cv2.ADAPTIVE_THRESH_GAUSSIAN_C

    thresh_1 = cv2.adaptiveThreshold(gray, 255, method, cv2.THRESH_BINARY, x, 2)  

    cv2_imshow(thresh_1)

adaptive_threshold(img)
```
