```python
import matplotlib.pyplot as plt
```

```python
def conv_output(w, k, p, s):
    return  ((w-k+2*p)/s) + 1

w = 32
k = 5
p = 1
s = 1

conv_output(w, k, p, s)
```

```python
outputs = []
w = 32
k = 5
p = 1
s = 1
w_sizes = [16, 32, 64]
for w in w_sizes:
    outputs.append(conv_output(w, k, p, s))
    
plt.plot(w_sizes, outputs)
```
