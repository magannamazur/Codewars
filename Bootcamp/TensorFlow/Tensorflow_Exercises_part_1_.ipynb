# Task 3 - differntiation

We try to calculate the derivative of$ f$ in $x$. Read the example:

image.png

```python
# Calculate the sinus derivative at 0

# use tensorflow's implementation of sinus function!!!
# https://www.tensorflow.org/api_docs/python/tf/math/sin
x = tf.Variable(0.0)

def f(x):
  return tf.sin(x)

with tf.GradientTape() as tape:
    y = f(x)

pochodna = tape.gradient(y,x)
pochodna
```
