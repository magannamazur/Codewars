```python
import matplotlib.pyplot as plt
# %matplotlib inline
import pandas as pd
import numpy as np
```

```python
## wczytaj plik temperature.csv bez ustawien
```

```python
df = pd.read_csv('Data/temperature.csv')
df
```

```python
## przefiltruj dane dla Xian oraz dla okresu mniejszego niz 1822-01-01
## zrob to dla zmiennej df_xian (może być w jednej linijce albo w dwóch :)
## za pomoca loca albo poprzez zwykle przypisanie series to series dokonaj zastapienia wartosci AverageTemperature i AverageTemperatureUncertainty
## wartościami skróconymi do jednego miejsca po kropce :) poprzez
## series.round https://pandas.pydata.org/docs/reference/api/pandas.Series.round.html
```

```python
df_xian = df[(df['City'] == 'Xian') & (df['dt'] < '1822-01-01')]
df_xian['AverageTemperature'] = df_xian['AverageTemperature'].round(1)
df_xian['AverageTemperatureUncertainty'] = df_xian['AverageTemperatureUncertainty'].round(1)
df_xian
```

```python
## wyświetl dwie linie na wykresie liniowym
## dla figsize 20 i 10
## ustaw tytul
## ustaw xlabel i xlabel
## zrob siatke za pomoca grida
## zrob rotacje w xtics
## wystaw markery na zgięciach
## linie powinny mieć różne kolory :) 
## wstaw tekst na zgięciach
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

ax.set_title("Nasz pierwszy wykres")

ax.set_xlabel("Czas", rotation=45, labelpad=20)
ax.set_ylabel("Sprzedaż", )
plt.xticks(rotation=45, ha='right')
plt.grid(True)

ax.plot(df_xian['dt'], df_xian['AverageTemperature'], marker="o", markersize=5)
ax.plot(df_xian['dt'], df_xian['AverageTemperatureUncertainty'], marker="o", markersize=5)

for i,j in zip(df_xian['dt'],df_xian['AverageTemperature']):
    ax.annotate(str(j),xy=(i,j))
    
for i,j in zip(df_xian['dt'],df_xian['AverageTemperatureUncertainty']):
    ax.annotate(str(j),xy=(i,j))

plt.show()
```
