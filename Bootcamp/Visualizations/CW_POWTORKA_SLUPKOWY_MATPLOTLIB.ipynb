```python
import matplotlib.pyplot as plt
# %matplotlib inline
import pandas as pd
import numpy as np
```

```python
## wczytaj plik population.csv bez ustawien
```

```python
df = pd.read_csv('Data/population.csv')
df
```

```python
## dla zmiennej sex wystaw unikatowe wartości w liście
## wyświetl series a nastepnie użyj metody unique 
## https://pandas.pydata.org/docs/reference/api/pandas.Series.unique.html
## następnie na liście użyj sortowania (możesz wpisać w googlach list sort for python)
```

```python
sex = list(df['Sex'].unique())
sex.sort()
sex
```

```python
## to powinno być w jednej linijcie

## dla zmiennej value wystaw zsumowane wartości po SEX
## wyświetl df dwoch kolumn
## dodaj do tego groupby po sex
## https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html
## dodaj do tego sum
## nastepnie zrob reset_index z wartoscia argumentu drop jako False 
## https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reset_index.html
## https://pandas.pydata.org/docs/reference/api/pandas.Series.unique.html
## przypisz wartości Value jako liste do zmiennej count poprzez wystawienie 
```

```python
df['Value'] = df['Value'].astype(float)

count = list(df[['Sex','Value']].groupby(by='Sex').sum().reset_index(drop=False).sort_values(by='Sex')['Value'].values)
count
```

```python
## przygotuj wykres slupkowy w matplotlibie z fig i ax jako subplots
## wstaw tytul jako Sex and counts
## wstaw ylabel jako count
## wstaw legende jako Sex color
## wyświetl wartości na słupkach
## wstaw trzy kolory i trzy labele na dole
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
fig, ax = plt.subplots()

sex = sex
counts = count
bar_labels = ['red', 'blue', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:orange']


ax.bar(sex, 
       counts, 
       label=bar_labels, 
       color=bar_colors)

for bars in ax.containers:
    ax.bar_label(bars, fmt = '%d')

ax.set_ylabel('Count')
ax.set_title('Sex and counts')
ax.legend(title='Sex color')
ax.ticklabel_format(style='plain', axis="y")
plt.show()
```
