```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = px.data.tips()
df
```

```python
sns.countplot(data=df, x='day', order=df['day'].unique())
plt.show()
```

```python
sns.countplot(data=df, x='day', order=df['day'].value_counts(ascending=True).index)
plt.show()
```

```python
sns.countplot(data=df, x='day', order=df['day'].value_counts(ascending=False).index)
plt.show()
```

```python
sns.countplot(data=df, x='day', hue='sex')
plt.show()
```

```python
ax = sns.countplot(data=df, x='day', hue='sex', color='aquamarine')
for label in ax.containers:
    ax.bar_label(label)
    
plt.show()
```
