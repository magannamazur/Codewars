```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(data=df, x='day')
plt.show()
```

```python
sns.countplot(data=df, x='day', order=df['day'].value_counts(ascending=False).index)
plt.show()
```

```python
sns.countplot(data=df, x='day', order=df['day'].value_counts(ascending=True).index)
plt.show()
```

```python
sns.countplot(data=df, x='day', hue='sex')
plt.show()
```

```python
sns.countplot(data=df, y='day', hue='sex')
plt.show()
```

```python
ax = sns.countplot(data=df, x='day', hue='sex', color='aquamarine')

for label in ax.containers:
    ax.bar_label(label)
    
plt.show()
```

```python
ax = sns.countplot(data=df, x='day', hue='sex', palette='Set2')

for label in ax.containers:
    ax.bar_label(label)
    
plt.show()
```

```python
penguins = sns.load_dataset("penguins")
sns.displot(penguins, x="flipper_length_mm", binwidth=3)
```

```python
tips = sns.load_dataset("tips")
sns.displot(tips, x="size", bins=[1,2,3,4,5,6])
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", element='step')
```

```python
sns.displot(penguins, x="flipper_length_mm", kind='kde')
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
g = sns.JointGrid(data=penguins, x="bill_length_mm", y="bill_depth_mm")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
```

```python
sns.pairplot(penguins)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", style="time")
```

```python
sns.set_theme()

tips = sns.load_dataset("tips")

sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)

plt.show()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day", col="time", row="sex")
plt.show()
```

```python
sns.heatmap(tips.corr(), annot=True)
plt.show()
```
