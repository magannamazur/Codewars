```python
df = pd.DataFrame({
    'Male': [35, 20, 43, 12, 19],
    'Female': [20, 11, 18, 6, 9]
 })

Class = ["Fisrt","Second","Third","Fourth","Fifth"]

ax = df.plot(stacked=True, kind = 'bar')

for bar in ax.patches: # mo≈ºna utozamiac z containsers, bar to jest ka≈ºdy osobny kolor s≈Çupka
    print(bar)
    height = bar.get_height()
    width = bar.get_width()
    
    x = bar.get_x() # wspo≈Çrzedne z poczatku slupka
    y = bar.get_y()
    
    label_text = height
    label_x = x + width/2
    label_y = y + height/2
    
    ax.text(label_x, label_y, label_text, ha='center', va= 'center') # nadanie warto≈õci wykresom na kolumnach

ax.set_xticklabels(Class, rotation = 'horizontal')    
plt.show()    
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Random test data
np.random.seed(19680801)
all_data = [np.random.normal(0, std, size=100) for std in range(1, 4)]
labels = ['x1', 'x2', 'x3']

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

# rectangular box plot
bplot1 = ax1.boxplot(all_data,
                     notch=False,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax1.set_title('Rectangular box plot')



# notch shape box plot
bplot2 = ax2.boxplot(all_data,
                     notch=True,  # notch shape
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks



ax2.set_title('Notched box plot')





# fill with colors
colors = ['pink', 'lightblue', 'lightgreen']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

        
        
# adding horizontal grid lines
for ax in [ax1, ax2]:
    ax.yaxis.grid(True)
    ax.set_xlabel('Three separate samples')
    ax.set_ylabel('Observed values')

plt.show()
```
